```python
#Cargar el dataset
import pandas as pd

df= pd.read_csv('SABERTYT20162.csv', sep=';', encoding="latin1")



#cortar dataset. Tomé todas las líneas con : y luego de la coma, se se adjunta al campo de las columnas
# 50 es por que toma desde el índice 50 hasta el final con :
df=df.iloc[:,50:]
```

```python
# ahora se dividen los datos en categóricos y numéricos
#numéricos
print("Variables Numéricas")
df_numéricas=df.select_dtypes(include=['float',"int"])
print(df_numéricas)
#categóricas
print("\nVariables Categóricas")
df_categóricas=df.select_dtypes(include=['object'])
print(df_numéricas)
```

```python
# Import
import numpy as np

#se importa in imputador
from sklearn.impute import SimpleImputer

# Pipeline
from sklearn.pipeline import Pipeline

# Scaler estandar preprocesado
from sklearn.preprocessing import StandardScaler

# Splitter
from sklearn.model_selection import train_test_split

# Importamos un clasificador: Support Vector Classifier para ML
from sklearn.svm import SVC

#importar otra librería desde chat
from sklearn.metrics import accuracy_score
```

```python
# Genera un problema de clasificación random de n-classes.

from sklearn.datasets import make_classification

#X, y = make_classification(random_state=0)
```

```python
# Dividimos en test y train
X_train, X_test, y_train, y_test = train_test_split(df_numéricas, df['MOD_COMUNI_ESCRITA_PGREF'] , random_state=0)

# Crea el pipeline
# El pipeline puede usarse como un estimador
# y evita data leaking del test set en el train set

pipe = Pipeline([
    ('scaler', StandardScaler()),
    #('imputer', SimpleImputer(strategy='mean'))
    ('svc', SVC())
    ])


# Se aplican los pasos en el dataset de entrenamiento
pipe.fit(X_train, y_train)

# se aplican los pasos en el dataset de prueba
pipe.score(X_test, y_test)
```

```python
#pruebas con chat
#Hacer predicciones en el conjunto de prueba
predictions = pipe.predict(X_test)

# Paso 5: Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy*100,"%")

# estos resultados apuntan que chat hace un código más largo que el el jose armando
#!!!!!NOTA: REVISAR CUÁL ES LA DIFERENCIA ENTRE AMBOS CÓDIGOS PORQUE LOS DOS PRESENTAN EL MISMO RESULTADO.!!!!!!!!
```
